```python
# Client-Server Connection, IP addresses and port numbers are used in networking to establish connections between devices, particularly in a client-server model.
```

```python
import socket

def create_http_response(content):
    response = f"""HTTP/1.1 200 OK
Content-Type: text/html

<html>
<head><title>Simple HTTP Server</title></head>
<body>
<h1>Server Response: {content}</h1>
<form method="POST">
    <label for="message">Enter a message:</label>
    <input type="text" id="message" name="message">
    <input type="submit" value="Submit">
</form>
</body>
</html>
"""
    return response.encode()

server_port = 80
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', server_port))
server_socket.listen(5)

print(f"Server listening on port {server_port}")

while True:
    client_socket, address = server_socket.accept()
    print(f"Connection from {address}")

    request_data = client_socket.recv(1024)
    print(f"Received request:\n{request_data.decode()}")

    user_input = ""
    if b"POST" in request_data:
        form_data_start = request_data.find(b'\r\n\r\n') + 4
        form_data = request_data[form_data_start:].decode()
        user_input = form_data.split('=')[1]

    response_data = create_http_response(user_input)
    client_socket.sendall(response_data)

    client_socket.close()
```
